```python
import pandas as pd
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus

# Database configuration
db_user = 'root'
db_password = 'Portal@77'
db_host = 'localhost'
db_name = 'BusTransport'

# Load the CSV file
csv_file_path = 'C:/Users/Dine24/Python Course/Guvi-Course/Cleaned_RedBus_Details.csv'
data = pd.read_csv(csv_file_path)

# Define table schema
table_creation_query = """
CREATE TABLE IF NOT EXISTS bus_routes (
    id INT AUTO_INCREMENT PRIMARY KEY,
    route_name TEXT NOT NULL,
    route_link TEXT NOT NULL,
    bus_name TEXT NOT NULL,
    bus_type TEXT NOT NULL,
    departing_time TIME NOT NULL,
    duration TEXT NOT NULL,
    reaching_time TIME NOT NULL,
    star_rating FLOAT,
    price DECIMAL(10, 2),
    seats_available INT
);
"""

# Initialize database connection
password_encoded = quote_plus(db_password)
engine = create_engine(f"mysql+pymysql://{db_user}:{password_encoded}@{db_host}/{db_name}")

# Create the table
try:
    with engine.connect() as connection:
        connection.execute(text(table_creation_query))  # Use text() for the query
        print("Table 'bus_routes' created successfully.")
except Exception as e:
    print(f"Error creating table: {e}")

# Preprocess the data for insertion
try:
    # Ensure 'seats_available' is numeric
    if 'seats_available' in data.columns:
        data['seats_available'] = pd.to_numeric(data['seats_available'], errors='coerce').fillna(0).astype(int)

    # Insert data into the table
    with engine.connect() as connection:
        data.to_sql('bus_routes', con=engine, if_exists='append', index=False)
        print("Data inserted successfully into 'bus_routes' table.")
except Exception as e:
    print(f"Error inserting data: {e}")
```
